```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.filters.filtertools import convolution_filter
import statsmodels.graphics.tsaplots as tsaplots
from statsmodels.tsa.seasonal import _extrapolate_trend
```

```python
# df = pd.read_csv('data/PRSA_Data_Aotizhongxin_20130301-20170228.csv') # 935
# df = pd.read_csv('data/PRSA_Data_Changping_20130301-20170228.csv') # 628
# df = pd.read_csv('data/PRSA_Data_Dingling_20130301-20170228.csv') # 730
# df = pd.read_csv('data/PRSA_Data_Dongsi_20130301-20170228.csv') # 663
# df = pd.read_csv('data/PRSA_Data_Guanyuan_20130301-20170228.csv') # 474
# df = pd.read_csv('data/PRSA_Data_Gucheng_20130301-20170228.csv') # 507
# df = pd.read_csv('data/PRSA_Data_Huairou_20130301-20170228.csv') # 980
# df = pd.read_csv('data/PRSA_Data_Shunyi_20130301-20170228.csv') # 1296
# df = pd.read_csv('data/PRSA_Data_Tiantan_20130301-20170228.csv') #1118
# df = pd.read_csv('data/PRSA_Data_Wanliu_20130301-20170228.csv') # 575
# df = pd.read_csv('data/PRSA_Data_Wanshouxigong_20130301-20170228.csv') # 669
df = pd.read_csv('data/PRSA_Data_Nongzhanguan_20130301-20170228.csv') # 446
df = df.dropna()
# df.isnull().sum().sort_values(ascending = False)

# sunspots = pd.read_csv('data/Sunspots.csv')
# sunspots = sunspots.set_index(pd.to_datetime(sunspots['Date'])) \
#     .drop(columns=['Unnamed: 0','Date']) \
#     .rename(columns={'Monthly Mean Total Sunspot Number':'values'})

df = df[['year','month','day','hour','SO2']]

df = df.set_index(pd.to_datetime(
    df.day.astype("str") + "-" + 
    df.month.astype('str') + "-" + 
    df.year.astype('str') + " " + 
    df.hour.astype('str') + ":0:0.0", 
    dayfirst=True, format='%d-%m-%Y %H:%M:%S.%f')
                 )
df = df.drop(columns=['year','month','day', 'hour'])
df.head()
df.info()
# y_hist = sunspots.loc['1900':'1984','values']
# y_future = sunspots.loc['1985':'2000','values']
# y_future
```

```python
y_hist = df.loc['2013-03-01T00:00:00.000000000':'2016-07-01T00:00:00.000000000','SO2']
y_future = df.loc['2016-07-02T00:00:00.000000000':'2017-02-28T23:00:00.000000000','SO2']

plt.figure(figsize=(16,5))
plt.plot(y_hist, 'k')
plt.plot(y_future,'k', alpha=0.5)
```

Get the trends

```python
periods = 300
flt = np.ones(periods+1)
flt[0] = 0.5
flt[-1] = 0.5
flt = flt / periods
sum(flt) # 0.999999999999997
```

```python
trend = convolution_filter(y_hist, flt)
trend = _extrapolate_trend(trend, periods + 1)

plt.figure(figsize=(16,5))
plt.plot(y_hist, 'k', label='SO2 Data')
plt.plot(trend, 'r', linewidth=3, label='Trend')
plt.plot(y_future, 'k', alpha=0.5)
plt.legend()
```

Get de-trended data

```python
detrended = y_hist - trend
detrended = detrended.to_numpy()

plt.figure(figsize=(16,5))
plt.plot(detrended)
```

Split the historical data into seasons (years)

```python
# 2013-7-01 00:00:00 , 2014-07-01 00:00:00, 2015-7-01 00:00:00, 2016-07-01 00:00:00
# 2697                 10808                19174               27642

low_indexes = [2697,10808,19174,27642]
lows = y_hist.index[low_indexes]
lows

# DatetimeIndex(['2013-07-01 01:00:00', '2014-07-01 00:00:00',
#                '2015-07-01 00:00:00', '2016-07-01 00:00:00'],
#               dtype='datetime64[ns]', freq=None)

# 2013-08-01 00:00:00 , 2014-09-01 00:00:00, 2015-08-01 00:00:00, 2016-07-01 00:00:00
# 3394                 12242                19911               27642


# low_indexes = [3395,12242,19911,27642]
# lows = y_hist.index[low_indexes]
# lows
```

```python
plt.figure(figsize=(16,5))
plt.plot(y_hist, 'k')
plt.plot(y_future, 'k', alpha=0.5)
plt.plot(trend, 'r', linewidth=3)
for low in lows:
    plt.axvline(x=pd.to_datetime(low), color='blue')
```

Split de-trended data into seasons

```python
period = int(np.round(np.mean(np.diff(low_indexes))))
number_seasons = len(lows) - 1

seasonals = np.empty((period, number_seasons))
for p in range(number_seasons):
    s = detrended[ low_indexes[p] : low_indexes[p]+period ]
    s = 2 * (s - np.min(s)) / (np.max(s) - np.min(s))
    seasonals[:,p] = s
    
mean_seasonals = seasonals.mean(axis=1)

plt.figure(figsize=(16,5))
for p in range(number_seasons):
    plt.plot(seasonals[:,p], color=0.5*np.ones(3))
plt.plot(mean_seasonals, color='blue', linewidth=3)
```

Smoth seasonal data

```python
filter_size = 100
fltr = np.repeat(1.0 / filter_size, filter_size)

seasonals = np.empty((period, number_seasons))
for p in range(number_seasons):
    s = detrended[ low_indexes[p] : low_indexes[p]+period ]
    s = convolution_filter(s, fltr)
    s = _extrapolate_trend(s, filter_size)
    s = (s-np.min(s)) / (np.max(s) - np.min(s))
    seasonals[:,p] = s
    
mean_seasonals = seasonals.mean(axis=1)

plt.figure(figsize=(16,5))
for p in range(number_seasons):
    plt.plot(seasonals[:,p], color=0.5*np.ones(3))
plt.plot(mean_seasonals, color='blue',linewidth=3)
```

Remove outliers

```python
seasonals_1 = seasonals[:,[0,1]]

mean_seasonals_1 = seasonals_1.mean(axis=1)

plt.figure(figsize=(16,5))
plt.plot(seasonals_1[0], color=0.5*np.ones(3), label='historical data')
plt.plot(seasonals_1, color=0.5*np.ones(3))
plt.plot(mean_seasonals_1, color='blue',linewidth=3)
```

```python
seasonals_2 = seasonals[:,[0,2]]

mean_seasonals_2 = seasonals_2.mean(axis=1)

plt.figure(figsize=(16,5))
plt.plot(seasonals_2[0], color=0.5*np.ones(3), label='historical data')
plt.plot(seasonals_2, color=0.5*np.ones(3))
plt.plot(mean_seasonals_2, color='blue',linewidth=3)
```

```python
seasonals_3 = seasonals[:,[1,2]]

mean_seasonals_3 = seasonals_3.mean(axis=1)

plt.figure(figsize=(16,5))
plt.plot(seasonals_3[0], color=0.5*np.ones(3), label='historical data')
plt.plot(seasonals_3, color=0.5*np.ones(3))
plt.plot(mean_seasonals_3, color='blue',linewidth=3)
```

Build seasonal template

```python
seasonal = pd.Series(index=y_hist.index, data=0)

for low in low_indexes:
    if low + period < len(seasonal):
        seasonal[ low : low+period ] = mean_seasonals
    else:
        seasonal[ low: ] = mean_seasonals[ :len(seasonal) - (low+period)]
        
seasonal = seasonal / np.max(seasonal)

plt.figure(figsize=(16,5))
plt.plot(seasonal)
```

Final model

```python
season_and_trend = 2 * trend * seasonal

plt.figure(figsize=(16,5))
plt.plot(y_hist,'k', label='data', linewidth=1)
plt.plot(trend, label='trend', color='red', linewidth=3)
plt.plot(season_and_trend,label='trend & seasonal', color='royalblue',linewidth=3)
plt.plot(y_future,'k', alpha=0.5, linewidth=1)
```

Model Residue

```python
residue = y_hist - season_and_trend

plt.figure(figsize=(14,4))
plt.plot(residue)
plt.grid()
```

Size of residue

```python
MAEm = np.abs(residue).mean()
RMSEm = np.sqrt( np.square(residue).mean() )
print(f'MAEm {MAEm}  RMSEm {RMSEm}')
```

Stationarity of residue

```python
_, ax = plt.subplots(figsize=(10,4))
tsaplots.plot_acf(residue,lags=20, ax=ax)
plt.show()
```

```python
Forecast
```

```python
f = len(y_future)
f_periods = int(1 + f/period)
yhat_trend = pd.Series(index=y_future.index, data=trend[-1])

yhat_seasonal = pd.Series(index=y_future.index)
for i in range(len(yhat_seasonal)):
    yhat_seasonal[i] = seasonal[-(f_periods*len(mean_seasonals)-i)]
    
yhat = 2 * yhat_trend * yhat_seasonal

plt.figure(figsize=(16,5))
plt.plot(y_hist,'k',label='data',linewidth=1)
plt.plot(trend,label='trend',color='red',linewidth=3)
plt.plot(season_and_trend,label='model',color='royalblue',linewidth=3)
plt.plot(yhat_trend,label='forecast trend',color='orange',linewidth=3)
plt.plot(yhat,label='forecast',color='limegreen',linewidth=3)
plt.plot(y_future,'k',alpha=0.5,linewidth=1)
```

Prediction error

```python
pred_error = y_future - yhat

plt.figure(figsize=(14,4))
plt.plot(residue,'k')
plt.plot(pred_error,'m')
```

Size of prediction error

```python
MAEp = np.abs(pred_error).mean()
RMSEp = np.sqrt( np.square(pred_error).mean() )
print(f'MAEp {MAEp}  RMSEp {RMSEp}')
```
